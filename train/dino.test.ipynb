```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import pandas as pd
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Mapeamento de classes para valores num√©ricos
classe_para_valor = {
    "CACTUS_SMALL": 0,
    "CACTUS_LARGE": 1,
    "PTERODACTYL": 2
}

acao_para_valor = {
    "JUMP": 0,
    "CROUCHED": 1,
    "RUN": 2,
}
```

```python
model = tf.keras.models.load_model('./dino_model/')
```

```python
df = pd.read_csv('./test.csv', skiprows=0)
input_data = df.iloc[:, 0:4].values
input_data = np.array([[classe_para_valor[classe], dist_x, dist_y, dist_z]
                      for classe, dist_x, dist_y, dist_z in input_data])
input_data[:, 0] = input_data[:, 0].astype(int)
```

```python
for i in range(len(input_data)):
    data_point = input_data[i]

    predicted_output = model.predict(np.array([data_point]))

    score = tf.nn.softmax(predicted_output[0])

    class_names = ["JUMP", "Crouched", "RUN"]

    print(class_names[np.argmax(score)], i, data_point)
    #  print("Result {} {:.2f}%".format(class_names[np.argmax(score)],100 * np.max(score)));
```
